```python
from AlgorithmImports import *
from QuantConnect.DataSource import *
import pandas as pd
from datetime import datetime, timedelta

# Initialize QuantBook
qb = QuantBook()

# Add the SPX index and then the SPXW options
spx = qb.AddIndex("SPX", Resolution.Minute).Symbol
spxw = qb.AddIndexOption(spx, "SPXW", Resolution.Minute).Symbol

# Define the date range for April 2023
start_date = datetime(2023, 3, 1)
end_date = datetime(2023, 4, 30)

# Get historical option chain data
# This requests data for all contracts that were "tradable" in that period based on the filter
# You would need to refine the filter to get specific strikes/expirations
historical_data = qb.OptionChainProvider.GetOptionContractList(spxw, start_date)

print(f"Number of available contracts on {start_date}: {len(historical_data)}")
print(historical_data[0])

```
